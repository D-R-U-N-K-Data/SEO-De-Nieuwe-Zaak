# Functions

```python
def backlink(dataframe):
    
    backlink_params = {
    '?key': 'b9fd485dbbcc0d31bc30e828806fd14c',
    'type': 'backlinks_refdomains',
    'target': '',
    'target_type': 'domain',
    'export_columns': 'backlinks_num',
    'display_limit': '1'
    }

    backlinklist = []
    endpoint_url = 'https://api.semrush.com/analytics/v1/'
    
    finalBacklinkData = dataframe.copy()
    
    finalBacklinkData[backlink_params['export_columns']] = 0
    
    for i in range(0, (len(finalBacklinkData))-48):
        try:
            backlink_params['target'] = finalBacklinkData.Dn[i]

            data = requests.get(semrush_call(backlink_params, endpoint_url))
            parsed_data = parse_response(data.content)
        
            conv_parsed_data = str(parsed_data)
            conv_parsed_data = conv_parsed_data.split("'")
            conv_parsed_data = int(conv_parsed_data[1])
            backlinklist.append(conv_parsed_data)
        except:
            backlinklist.append(None)
    
    # Hash the backlink list and save it
    backlinklist_hashed = hash_variable(backlinklist)
    backlinklist_df = pd.DataFrame(backlinklist)
    backlinklist_df.to_csv(f"calls/backlinklist-{backlinklist_hashed}.csv")
    
    return backlinklist

def inbound(dataframe):
    
    inbound_params = {
    '?key': 'b9fd485dbbcc0d31bc30e828806fd14c',
    'type': 'backlinks',
    'target': '',
    'target_type': 'url',
    'export_columns': '',
    'display_limit': '1'
    }
    
    inboundlist = []
    endpoint_url = 'https://api.semrush.com/analytics/v1/'

    finalOutboundData = dataframe.copy()
    
    inbound_params['export_columns'] = 'internal_num'
    finalOutboundData['internal_num'] = 0

    for i in range(0, (len(finalOutboundData)-48)): 
        try:
            inbound_params['target'] = finalOutboundData.Ur[i]

            data = requests.get(semrush_call(inbound_params, endpoint_url))
            parsed_data = parse_response(data.content)
        
            conv_parsed_data = str(parsed_data)
            conv_parsed_data = conv_parsed_data.split("'")
            conv_parsed_data = int(conv_parsed_data[1])
            inboundlist.append(conv_parsed_data)
        except:
            inboundlist.append(None)
            
    # Hash the inbound list and save it
    inboundlist_hashed = hash_variable(inboundlist)
    inboundlist_df = pd.DataFrame(inboundlist)
    inboundlist_df.to_csv(f"calls/inboundlist-{inboundlist_hashed}.csv")
            
    return inboundlist

def outbound(dataframe):
    
    outbound_params = {
    '?key': 'b9fd485dbbcc0d31bc30e828806fd14c',
    'type': 'backlinks',
    'target': '',
    'target_type': 'url',
    'export_columns': '',
    'display_limit': '1'
    }
    
    outboundlist = []
    endpoint_url = 'https://api.semrush.com/analytics/v1/'

    finalOutboundData = dataframe.copy()
    
    outbound_params['export_columns'] = 'external_num'
    finalOutboundData['external_num'] = 0

    for i in range(0, (len(finalOutboundData)-48)): 
        try:
            outbound_params['target'] = finalOutboundData.Ur[i]

            data = requests.get(semrush_call(outbound_params, endpoint_url))
            parsed_data = parse_response(data.content)
        
            conv_parsed_data = str(parsed_data)
            conv_parsed_data = conv_parsed_data.split("'")
            conv_parsed_data = int(conv_parsed_data[1])
            outboundlist.append(conv_parsed_data)
        except:
            outboundlist.append(None)
            
    # Hash the outbound list and save it
    outboundlist_hashed = hash_variable(outboundlist)
    outboundlist_df = pd.DataFrame(outboundlist)
    outboundlist_df.to_csv(f"calls/outboundlist-{outboundlist_hashed}.csv")        
                
    return outboundlist
```
